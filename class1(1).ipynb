Write a code that takes a string as input and returns True if the string is a palindrome and False otherwise. A palindrome is a word, phrase, or sequence of characters that reads the same backward as forward. Ignore spaces, punctuation, and capitalization when checking for a palindrome.

```python
def is_palindrome(input_str):
    # Remove spaces and convert to lowercase
    cleaned_str = ''.join(char.lower() for char in input_str if char.isalnum())

    # Check if the cleaned string is equal to its reverse
    return cleaned_str == cleaned_str[::-1]

# Example usage:
input_string = "A man, a plan, a canal, Panama!"
result = is_palindrome(input_string)

print(result)
```

write a code that takes a list of numbers and returns the average using a for loop

```python
def cal_average(numbers):
    if not numbers:
        return 0
    total = 0
    count = 0

    for number in numbers:
        total += number
        count += 1

    average = total / count
    return average


number_list = [1, 2, 3, 4, 5]
result = cal_average(number_list)
print("The average is:", result)
```

Write a Python function that takes two lists of numbers as input and returns a new list containing common elements between the two lists.

```python
def find_common_elements(list1, list2):
  set1 = set(list1)
  set2 = set(list2)
  common_elements = set1.intersection(set2)
  return list(common_elements)

list1 = [1, 3, 5, 7, 9]
list2 = [2, 4, 6, 7, 10]
common_list = find_common_elements(list1, list2)
print(f"Common elements between {list1} and {list2}: {common_list}")
```

```python
##bar plot(to compare 2 datas)

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color="g")
plt.title("bar graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
plt.savefig("bar_graph.png")
```

**Seaborn**

```python
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
#create a scatter plot of total bill vs. tip amount

sns.scatterplot(x ="total_bill",y="tip",data=df)
plt.title("comparing")
```

```python
# create a bar plot of average tip of amount by day of the week

sns.barplot(x="day",y="tip",data=df)
```

```python
#histogram
sns.histplot(x="total_bill", data=df,kde=True)#kernel distribution estimation-kde
```

```python
#Box plot of total bill amounts by time of day

sns.boxplot(x="time",y="total_bill",data=df)
```

```python
# create a heatmap of correlation between numerical variables

sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
df.corr()
```

```python
#count plot of the no. of tips by day of the week

sns.countplot(x="sex",data=df
              )
```

```python
#pair plot - similar to subplot, we can see all graphs

sns.pairplot(data=df)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df)
```

```python
#joint plot of total bill and tip amount
sns.jointplot(x="total_bill",y="tip",hue="smoker",data=df)
```

```python
#linear regression plot of total bill and tip amoount
sns.lmplot(x="total_bill",y="tip",data=df)
```

```python
# create violin plot of total bill amount by day of the week

sns.violinplot(x="day",y="total_bill",data=df)
```

```python
#boxen plot
sns.boxenplot(x="day",y="total_bill",data=df)
```

**2nd Dataset**

```python
d=sns.load_dataset("iris")
d
```

```python
#Display the first few rows of the dataset
d.head(5)
```

```python
#Create a pair plot to visualize the relationships between sepal length, sepal width, petal length, and petal width.

sns.pairplot(data=d,hue="species")
plt.suptitle("pair plots")
```

```python
#Create a box plot to show the distribution of sepal lengths for each species

sns.boxplot(data=d,x="sepal_length",y="species")
```

```python
#Create histograms to visualize the distribution of sepal length, sepal width, petal length, and petal width for each species.
plt.subplot(2,2,1)
sns.histplot(data=d,x="sepal_length",hue="species",kde=True)
plt.title("Histogram of sepal length")

plt.subplot(2,2,2)
sns.histplot(data=d,x="sepal_width",hue="species",kde=True)
plt.title("Histogram of sepal width")

plt.subplot(2,2,3)
sns.histplot(data=d,x="petal_length",hue="species",kde=True)
plt.title("Histogram of petal length")

plt.subplot(2,2,4)
sns.histplot(data=d,x="petal_width",hue="species",kde=True)
plt.title("Histogram of petal width")
```

```python
#Create a correlation heatmap
sns.heatmap(d.corr(),annot=True,cmap="coolwarm")
d.corr()
```
